```python
import numpy as np
import cv2
from tensorflow.keras.models import load_model
import matplotlib.pyplot as plt

# Load the saved model
model = load_model('/content/drive/Shareddrives/MobileNetV2/Xception/best_ResNet101V2_model.h5')

# Define the image size expected by the model
img_size = (224, 224)

# Load and resize the input image using OpenCV
img = cv2.imread('/content/drive/Shareddrives/MobileNetV2/Inferencing/Healthy/2.jpg')
img = cv2.resize(img, img_size)

# Reshape the input data to match the expected input shape of the model
img = np.expand_dims(img, axis=0)

# Make predictions on the input data
predictions = model.predict(img)

# Get the predicted class label
predicted_class = np.argmax(predictions, axis=1)[0]

class_names = ['Rust', 'Healthy']

# Show the input image and the predicted class label
plt.imshow(cv2.cvtColor(img[0], cv2.COLOR_BGR2RGB))
plt.title(class_names[predicted_class])
plt.show()
```

```python
import tensorflow as tf
from sklearn.metrics import classification_report

# Load the saved model
model = tf.keras.models.load_model('/content/drive/Shareddrives/MobileNetV2/Xception/best_ResNet50V2_model.h5')

# Load the test data
test_data = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255).flow_from_directory(
    '/content/drive/Shareddrives/MobileNetV2/Dataset_withval/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Predict the class probabilities for the test data
y_pred = model.predict(test_data)

# Convert the class probabilities to class labels
y_pred_labels = tf.argmax(y_pred, axis=1)
y_true_labels = test_data.classes

# Get the classification report
report = classification_report(y_true_labels, y_pred_labels, target_names=test_data.class_indices.keys())

# Print the classification report
print(report)
```

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix

# Load the saved model
model = tf.keras.models.load_model('/content/drive/Shareddrives/MobileNetV2/Xception/best_ResNet50V2_model.h5')

# Load the test data
test_data = tf.keras.preprocessing.image.ImageDataGenerator(rescale=1./255).flow_from_directory(
    '/content/drive/Shareddrives/MobileNetV2/Dataset_withval/test',
    target_size=(224, 224),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

# Predict the class probabilities for the test data
y_pred = model.predict(test_data)

# Convert the class probabilities to class labels
y_pred_labels = np.argmax(y_pred, axis=1)
y_true_labels = test_data.classes

# Get the confusion matrix
cm = confusion_matrix(y_true_labels, y_pred_labels)

# Plot the confusion matrix
sns.heatmap(cm, annot=True, cmap='Blues')
plt.xlabel('Predicted Labels')
plt.ylabel('True Labels')
plt.show()
```
