```python
import os
import csv

# Set the directory path
dir_path = "/content/drive/MyDrive/Dataset/Photos _Clean"

# Set the path to the CSV file
csv_file_path = "/content/drive/MyDrive/Dataset/RoCoLe-classes_Clean.csv"

label_to_delete =  "red_spider_mite"

# Open the CSV file and read the contents
with open(csv_file_path, "r",) as csvfile:
    reader = csv.reader(csvfile)

    # Iterate over each row in the CSV file
    for i, row in enumerate(reader):
        # Skip the first row (header)
        if i == 0:
            continue

        filename, _, label  = row

        # Check if the label matches the label to delete
        if label == label_to_delete:
            # Construct the full path to the file
            file_path = os.path.join(dir_path, filename)

            # Check if the file exists
            if os.path.exists(file_path):
                # Delete the file
                os.remove(file_path)
                print(f"File {filename} deleted.")
            else:
                print(f"File {filename} not found.")
        else:
            print(f"Skipping file {filename} with label {label}.")
```
