```python
  import numpy as np
  import cv2
  from tensorflow.keras.models import load_model
  import matplotlib.pyplot as plt

# Load the saved model
model = load_model('/content/drive/Shareddrives/MobileNetV2/Xception/best_ResNet101V2_model.h5')

# Define the image size expected by the model
img_size = (224, 224)

# Load and resize the input image using OpenCV
img = cv2.imread('/content/drive/Shareddrives/MobileNetV2/Inferencing/Healthy/1.jpg')
img = cv2.resize(img, img_size)

# Reshape the input data to match the expected input shape of the model
img = np.expand_dims(img, axis=0)

# Make predictions on the input data
predictions = model.predict(img)

# Get the predicted class label
predicted_class = np.argmax(predictions, axis=1)[0]

class_names = ['Healthy', 'Rust']

# Show the input image and the predicted class label
plt.imshow(cv2.cvtColor(img[0], cv2.COLOR_BGR2RGB))
plt.title(class_names[predicted_class])
plt.show()
```
